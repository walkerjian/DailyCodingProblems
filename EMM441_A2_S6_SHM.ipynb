# Simple Harmonic Motion (SHM)
Simple Harmonic Motion (SHM) is a type of periodic motion where the restoring force is directly proportional to the displacement and acts in the direction opposite to that of displacement. It is at the very heart of most mathematical physics.

We will start by deriving the differential equation for SHM and then solve it step by step.

### 1. Deriving the Differential Equation

Consider a mass $ m $ attached to a spring. Let $ x(t) $ be the displacement of the mass from its equilibrium position at any time $ t $. According to Hooke's Law, the force exerted by the spring is proportional to the displacement and opposite in direction.

Mathematically, this can be expressed as:

$ F = -kx $

where $ k $ is the spring constant, and $ x $ is the displacement from the equilibrium position.

Newtonâ€™s second law states that the force acting on an object is equal to the mass of the object multiplied by its acceleration:

$ F = ma $

Acceleration $ a $ can be expressed as the second derivative of displacement with respect to time (since acceleration is the rate of change of velocity, which is itself the rate of change of displacement):

$ a = \frac{d^2x}{dt^2} $

Combining the equations:

$ ma = -kx $
$ m \frac{d^2x}{dt^2} = -kx $

Dividing by $ m $, we get:

$ \frac{d^2x}{dt^2} = -\frac{k}{m} x $

Define $ \omega^2 = \frac{k}{m} $. (WHY?)

The differential equation becomes:

$ \frac{d^2x}{dt^2} = -\omega^2 x $

**This is the standard form of the differential equation for SHM.**

### 2. Solving the Differential Equation

We will solve this differential equation:

$ \frac{d^2x}{dt^2} = -\omega^2 x $

#### Step 1: Guess the Solution
We **guess that the solution could be of the form**

$ x(t) = A \cos(\omega t + \phi) $,

 where $ A $ is the amplitude, $ \omega $ is the angular frequency, and $ \phi $ is the phase constant.

#### Step 2: Verify the Solution
To verify, differentiate $ x(t) $ twice with respect to $ t $ (exercise):

$ \frac{dx}{dt} = -A\omega \sin(\omega t + \phi) $

$ \frac{d^2x}{dt^2} = -A\omega^2 \cos(\omega t + \phi) $

Substituting $ \frac{d^2x}{dt^2} $ back into our differential equation:

$ -A\omega^2 \cos(\omega t + \phi) = -\omega^2 (A \cos(\omega t + \phi)) $

which is true, so we confirm the guess is a valid solution.

#### Step 3: General Solution
The general solution to the differential equation of SHM is:

$ x(t) = A \cos(\omega t + \phi) $

where:
- $ A $ (amplitude) is the maximum displacement from the equilibrium.
- $ \omega $ (angular frequency) is $ \sqrt{\frac{k}{m}} $.
- $ \phi $ (phase constant) depends on initial conditions.

### Understanding the Parameters:
- The **period** $ T $ of the motion is given by $ T = \frac{2\pi}{\omega} $.
- The **frequency** $ f $ is $ f = \frac{1}{T} = \frac{\omega}{2\pi} $.

# Experiment to Investigate SHM: Mass-Spring System

A simple yet effective experiment to investigate Simple Harmonic Motion (SHM) is to use a mass-spring system.

#### Materials Needed:
- A spring with a known spring constant $ k $. (we need to find this too!)
- A mass $ m $ that can be attached to the spring.
- A stand from which to suspend the spring.
- A ruler or some measuring device to measure displacement.
- A stopwatch to measure the period of oscillation.

#### Experimental Setup:
1. **Setup the Spring**: Attach one end of the spring to the stand and the other end to the mass.
2. **Initial Displacement**: Pull the mass down to stretch the spring by a small, measured displacement from the equilibrium position and release it from rest.
3. **Measurement of Oscillations**:
   - Measure the maximum amplitude $ A $ of the oscillations.
   - Use the stopwatch to time several complete oscillations (at least 10 to improve accuracy), and then calculate the average period $ T $ of one oscillation.

#### Data Recording:
Record the displacement and the corresponding time period. This data will help in verifying the relationship between the period $ T $, the mass $ m $, and the spring constant $ k $ as derived theoretically:

$ T = 2\pi \sqrt{\frac{m}{k}} $

### Learning Points
- Observe how the period $ T $ changes with variations in $ m $ and $ k $. According to the theory, $ T $ should increase with mass and decrease with a stiffer spring (higher $ k $).
- The displacement plot shows how the mass moves back and forth around the equilibrium position, a characteristic of SHM.

This setup provides a hands-on understanding of the dynamics of SHM, directly linking theory with practical observation and simulation.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Constants
m = 0.5  # mass in kilograms
k = 20   # spring constant in N/m
A = 0.1  # amplitude in meters

# Derived quantities
omega = np.sqrt(k/m)  # angular frequency
T = 2 * np.pi / omega  # period

# Time array
t = np.linspace(0, 3*T, 300)

# Position as a function of time
x = A * np.cos(omega * t)

# Create the figure and axis
fig, ax = plt.subplots()
line, = ax.plot(t, x, label='x(t) = A*cos(Ï‰t)')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Displacement (m)')
ax.legend()

# Animation function
def animate(i):
    line.set_ydata(A * np.cos(omega * (t - 0.05 * i)))  # update the data
    return line,

# Create animation
ani = FuncAnimation(fig, animate, frames=200, interval=50, blit=True)

plt.show()
```
