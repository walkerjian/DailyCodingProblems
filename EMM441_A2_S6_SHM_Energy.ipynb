# Mathematically Modelling Energy in Simple Harmonic Motion (SHM)
In Simple Harmonic Motion (SHM), energy conservation is very useful for understanding how energy transforms between kinetic and potential forms without any net loss.

### Total Mechanical Energy in SHM

The total mechanical energy $ E $ in a mass-spring system undergoing SHM consists of kinetic energy $ K $ and potential energy $ U $ thus:

1. **Kinetic Energy (K):** This is the energy due to the motion of the mass. It is given by:

   $
   K = \frac{1}{2} mv^2
   $


   where $ v $ is the velocity of the mass.

2. **Potential Energy (U):** This is the energy stored in the spring when it is compressed or stretched. It is given by:

   $
   U = \frac{1}{2} kx^2
   $

   where $ x $ is the displacement from the **equilibrium** position. (DEFINE)

### Deriving the Energy Equations

The solution for **displacement** in SHM:
$ x(t) = A \cos(\omega t + \phi) $

The velocity $ v $ of the mass is the first derivative of displacement:
$ v(t) = \frac{dx}{dt} = -A\omega \sin(\omega t + \phi) $

Substitute these into the energy formulas:

- **Kinetic Energy**:

  $
  K = \frac{1}{2} m (-A\omega \sin(\omega t + \phi))^2 = \frac{1}{2} mA^2\omega^2 \sin^2(\omega t + \phi)
  $


- **Potential Energy**:

  $
  U = \frac{1}{2} k (A \cos(\omega t + \phi))^2 = \frac{1}{2} kA^2 \cos^2(\omega t + \phi)
  $


### Total Energy Conservation

Using the relationship

$ \omega^2 = \frac{k}{m} $,

 we can see the total energy is:

$ E = K + U $

$ E = \frac{1}{2} mA^2\omega^2 \sin^2(\omega t + \phi) + \frac{1}{2} kA^2 \cos^2(\omega t + \phi) $

$ E = \frac{1}{2} kA^2 (\sin^2(\omega t + \phi) + \cos^2(\omega t + \phi)) $

$ E = \frac{1}{2} kA^2 $

since $ \sin^2(\theta) + \cos^2(\theta) = 1 $.

### Python Simulation Demonstrating Energy Conservation
This Python code computes the displacement and velocity at various points in time and subsequently calculates the kinetic and potential energies. It then plots these energies against time to demonstrate:
- Kinetic energy is highest when the potential energy is lowest (and vice versa).
- The total energy remains constant, illustrating the principle of energy conservation in SHM.




```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
m = 0.5  # mass in kilograms
k = 20   # spring constant in N/m
A = 0.1  # amplitude in meters

# Derived quantities
omega = np.sqrt(k/m)  # angular frequency
T = 2 * np.pi / omega  # period

# Time array
t = np.linspace(0, 3*T, 300)

# Displacement and velocity
x = A * np.cos(omega * t)
v = -A * omega * np.sin(omega * t)

# Energies
K = 0.5 * m * v**2
U = 0.5 * k * x**2
E = K + U

# Plotting
plt.figure(figsize=(10, 8))
plt.plot(t, K, label='Kinetic Energy')
plt.plot(t, U, label='Potential Energy')
plt.plot(t, E, 'k--', label='Total Energy (constant)')
plt.xlabel('Time (s)')
plt.ylabel('Energy (Joules)')
plt.title('Energy in Simple Harmonic Motion')
plt.legend()
plt.grid(True)
plt.show()
```
